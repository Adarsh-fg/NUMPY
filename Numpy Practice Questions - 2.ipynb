#NumPy Array Copy and View

```python
import numpy as np
```

1. Create a NumPy array and make a copy of it. Modify the original array and print both
the original and copied arrays.

```python
a=np.array([1,2,3,4,5])
b=a.copy()

a[2]=101

print(a)
print(b)
```

2. Create a view of a NumPy array. Modify the view and print both the original and the
view. How do the results compare to using a copy?

```python
a=np.array([1,2,3,4,5])

b=a.view()

b[2]=101

print(a)
print(b)
```

3. Explain the difference between array.copy() and array.view() in NumPy. Provide an
example to illustrate the difference.

```python
a=np.array([1,2,3,4,5])
print("Original array ",a)

b=a.copy()
b[2]=101
print("Copy array after value changing is  ",b)
print("Original array after value changing ",a)
print()


c=a.view()
c[2]=1001
print("View array after value changing     ",c)
print("Original array after value changing ",a)
```

4. Create a 2D NumPy array and generate a view that represents only the first row.
Modify this view and print the changes in both the original array and the view.

```python
a=np.array([[1,2,3],[4,5,6]])
print(a)
print("----------------------")

b=a.view()
print(b[0])
print("----------------------")

b[0,1]=9
print(b[0])
print("----------------------")
print(b)
```

#Check Copy or View

5. Create a NumPy array and another array by slicing the first array. Check whether the
second array is a copy or a view of the first array.

```python
a=np.array([1,2,3,4,5])

b=a[0:3]
print(b)
print("----------------------")
print(b.base)   # so it is not copy it is a view
```

6. Use the np.may_share_memory() function to check if two arrays share the same
memory. Provide an example.

```python
a=np.array([1,2,3,4,5,6])
b=np.array([1,2,3,4,5,6])
x=np.may_share_memory(a,b)
print(x)
```

```python
a=np.array([1,2,3,4,5,6])
b=a.view()
x=np.may_share_memory(a,b)
print(x)

#if they share same memory it returns true
```

7. Create a NumPy array, then use the reshape() function on it. Check whether the
reshaped array is a view or a copy.

```python
a=np.array([1,2,3,4,5,6])
b=a.reshape(2,3)
print(b.base)          #it is a view
```

#NumPy Array Shape

8. Create a NumPy array and print its shape. Then reshape it to another valid shape and
print the new shape.

```python
a=np.array([1,2,3,4,5,6])
print("Original shape : ",a.shape)
b=a.reshape(2,3)
print("Reshaped shape :",b.shape)
```

9. Given a 1D array of length 12, reshape it into a 2D array with 3 rows. Then, reshape it
into a 3D array. Print the shapes at each step.

```python
a=np.arange(12)
print(a)
print("----------------------")
b=a.reshape(3,4)
print(b)
print("----------------------")
print("Shape :",b.shape)
print("----------------------")

c=a.reshape(3,2,-1)
print(c)
print("----------------------")
print("Shape :",c.shape)
print("----------------------")
```

10.Create a 3x4 matrix and then reshape it into a 2x6 matrix. Verify if reshaping changes
the data content.

```python
a=np.arange(12).reshape(3,4)
print(a)
print("----------------------")
print("Shape :",a.shape)
print("----------------------")

b=a.reshape(2,6)
print(b)
print("----------------------")
print("Shape :",b.shape)
```

#NumPy Array Reshaping

11.Reshape a 2D array into a 3D array. Then reshape it back to its original 2D shape.

```python
a=np.arange(12).reshape(3,4)
print(a)
print("----------------------")
b=a.reshape(2,2,-1)
print(b)
print("----------------------")
print("Shape :",b.shape)
print("----------------------")
a=b.reshape(3,4)
print(a)
print("----------------------")
print("Shape :",a.shape)
print("----------------------")
```

12.Given a 4x4 matrix, reshape it into a 2D array with dimensions 2x8. Then, reshape it
back to 4x4.

```python
a=np.arange(16).reshape(4,4)
print(a)
print("----------------------")
b=a.reshape(2,8)
print(b)
print("----------------------")
c=b.reshape(4,4)
print(c)
print("----------------------")
```

# Iterating with Different Data Types


20. Create a 2D integer array and use nditer() to iterate over it as if it were an float array.
Print the elements.


```python
a=np.arange(6).reshape(2,3)
for i in np.nditer(a,flags=['buffered'],op_dtypes=[float]):
  print(i)
```

21. Use nditer() to iterate over a mixed data type array, where some elements are floats

# Iterating with Different Step Sizes

23.Create a 1D array of 10 elements and iterate over it using a step size of 2. Print the
accessed elements.


```python
a=np.arange(10)
for i in np.nditer(a[::2]):
  print(i)
```
