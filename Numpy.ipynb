##EYE TOOL

**The eye tool returns a 2d array with 1s as the diagonal and 0 s elsewhere** .The diagonal can be main upper and lower depending on the parameter .**A positive k is used for the upper diagonal and negative k is for the lower and 0k is for the main diagonal**

```python
#python programing illustarting
import numpy as np

#metrix with 1 in diagonal
b=np.eye(2)
print("2x2\n",b)

b=np.eye(3,5)
print("\n3x5\n",b)

#matrix with r=4 c=5 and 1 in diagonal
#below main diagonal
b=np.eye(4,5, k=-1)    #default value is 0
print("\n4x5   (-1)\n",b)

#default value of k is 0

#above main diagonal
b=np.eye(4,5, k=1)
print("\n4x5   (1)\n",b)
```

# Numpy non zero

nonzero() returns the indices of the elemnts that are non zero

```python
#get the values and indices of the elemnt which is bigger then 10 in given array

import numpy as np

x=np.array([[0,10,20],[20,30,40]])

print("Array :\n",x)

#filering values bigger than 10

print("Values which are bigger than 10 are :",x[x>10])

#indices of thr elemnts that are bigger than 10 using np.nonzero()

print("Their indices are :",np.nonzero(x>10))

#output : array([row]),array([column])
```

This means :

*   20 at index(0,2)
*   20 at index(1,0)
*   30 at index(1,1)
*   40 at index(1,2)



```python
import numpy as np

x=np.array([[106,107,99,67,110],[26,76,96,200,101],[1,2,202,75,60]])

print("Array :\n",x)

print("values which are bigger then 100 are ",x[x>100])

print("Their indices are ",np.nonzero(x>100))
```

##  Discrete distribution

A discrete distribution deals with discrete random variables, which are variables that can take on only specific, distinct values. These values are countable, and there are gaps between each possible value.
* **Key Features of Discrete Distributions:**
  * The random variable can take only specific, separated values (e.g., 0, 1, 2, ...).
  * The probability of each individual outcome can be calculated, and the sum of all probabilities equals 1.
  * Common examples include counting the number of successes in trials (e.g., coin flips, number of defects in products).

## Continuous distribution

A continuous distribution deals with continuous random variables, which can take any value within a given range. These variables are not countable because there are infinitely many possible values between any two points in the range.
* Key Features of Continuous Distributions:
    * The random variable can take any value within a given range (e.g., any number between 0 and 1. including all decimals).
    * Probabilities are not assigned to individual points but to ranges of values. The probability of the random variable taking an exact value is always 0.
    * Instead of individual probabilities, continuous distributions are described by a **Probability Density Function (PDF)**.

##NumPy GCD Greatest Common Denominator
* The GCD (Greatest Common Denominator), also known as HCF (Highest Common Factor) is the biggest number that is a common factor of both of the numbers.

```python
import numpy as np

n1 = 6
n2 = 9
x = np.gcd(n1, n2)
print(x)
```

*Returns: 3 because that is the highest number both numbers can be divided by (6/3=2 and 9/3=3).*

* the np.gcd() function is used to compute GCD of num1 , num2 the gcd is the largest integer  that divides both numbers without leaving a remainder
* In this case , thr GCD of 6 and 9 is 3 beacause 3 is the largest number that divides both 6 and 9.

**Explanation of GCD**

* Explanation of GCD (Greatest Common Denominator):
  * For 6 and 9:
    * The divisors of 6: 1, 2, 3, 6
    * The divisors of 9: 1, 3, 9  
    * The common divisors: 1 and 3
    * The largest common divisor is 3, which makes the GCD of 6 and 9 equal to 3.

**Finding GCD in Arrays**

* To find the Highest Common Factor of all values in an array, you can use the **reduce()** method.
* **The reduce() method will use the ufunc, in this case the gcd() function, on each element, and reduce the array by one dimension.**

```python
# Find the GCD for all of the numbers in the following array:

import numpy as np

arr = np.array([20, 8, 32, 36, 16])

x = np.gcd.reduce(arr)

print(x)
```

Returns 4 because that is the highest number all values can be divided by.

* the function **np.gcd.reduce(arr)** computes the GCD of the entire by applying the GCD function cumulatively across all the elments in the array.
* The reduce function applies the GCD opertaion iteratively:
  * first finds the GCD of the first two numbers (gcd(20, 8) = 4)
  * Then, it takes this result and applies the GCD with the next gegr (ged(4,324), and so on.
  * This process continues **until all the numbers in the array are reduced to their greatest common divisor.**

# NumPy Set Operation

**What is a Set**

* A set in mathematics is a collection of unique elements
* Sets are used for operations involving frequent ***intersection, union and difference operations.***

**Create Sets in NumPy**
* We can use NumPy's unique() method to find unique elements from any array. Eg. create a set array, but remember that the set arrays should only be 1-D arrays.

```python
#convert following array with repeated elements to a set.

import numpy as np

arr=np.array([1,1,1,2,3,4,5,5,6,7])
x=np.unique(arr)            # no repeatation in set
print(x)
```

**Finding Union**

* To find the unique values of two arrays, use the **union1d()** method

```python
# Find union of the following two set arrays:

import numpy as np

arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([3, 4, 5, 6])

newarr = np.union1d(arr1, arr2)

print(newarr)
```

**Finding Intersection**

* To find only the values that are present in both arrays, use the intersect1d() method.

```python
# Find intersect of the following two set arrays:

import numpy as np

arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([3, 4, 5, 6])

newarr = np.intersect1d(arr1, arr2,assume_unique=True)

print(newarr)
```

Note: the intersect1d() method takes an optional argument assume_unique, which if set to True can speed up computation. It should always be set to True when dealing with sets.

**Finding Difference**  


* To find only the values in the first set that is NOT present in the seconds set, use the setdiff1d() method.

```python
# Find difference of the following two set arrays:

import numpy as np

arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([3, 4, 5, 6])

newarr = np.setdiff1d(arr1, arr2)

print(newarr)
```

```python
# Find difference of the following two set arrays:

import numpy as np

arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([3, 4, 5, 6])

newarr = np.setdiff1d(arr2,arr1)

print(newarr)
```

Note: the setdiff1d() method takes an optional argument assume_unique, which if set to True can speed up computation. It should always be set to True when dealing with sets.

**Finding Symmetric Difference**

* To find only the values that are NOT present in BOTH sets, use the setxor1d() method.

```python
# Find symmetric difference of the following two set arrays:

import numpy as np

arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([3, 4, 5, 6])

newarr = np.setxor1d(arr1, arr2)

print(newarr)
```

Note: the setxor1d() method takes an optional argument assume_unique, which if set to True can speed up computation. It should always be set to True when dealing with sets.
